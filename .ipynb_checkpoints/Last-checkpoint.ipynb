```python
#%%writefile cleaning.py
import pandas as pd
import numpy as np

#cleaning data def
def cleaning(all_data): #from data_df
    df = all_data.drop_duplicates()
    pd.set_option('display.max_colwidth', None)
    #for price
    if 'price' in df.columns:
        try:
            df['price'] = df['price'].str[:-3]
            df['price'] = df['price'].str.replace('\xa0', '')
            #more settings - in developing
            try:
                df['price'] = df['price'].map(int)
                if len(df.loc[df['price'] < 100, 'price'])>0:
                    df.loc[df['price'] < 100, 'price'] =0
            except:
                print("can't make price int")
            '''recover = df[~df['price'].str.isdigit()]
            if len(recover) > 0:
                df.loc[~leie['price'].str.isdigit(), ['price']] = recover['square_metre']
                df.loc[~leie['price'].str.isdigit(), ['r_type_amount']] = recover['price']
                df.loc[~leie['price'].str.isdigit(), ['square_metre']] = '0'
                df.loc[~leie['price'].str.isdigit(), ['price']] = leie['price'].str[:-3]
                df.loc[~leie['price'].str.isdigit(), ['price']] = leie['price'].str.replace('\xa0', '')
            df.loc[0, 'link'] = leie.loc[0, 'trash2']'''
        except:
            print('problem with price descr')
    #for price_desc
    if 'price_desc' in df.columns:
        try:
            df[['total_price', 'fellesutg']] = df['price_desc'].str.split(' ∙ ', expand=True)
            del df['price_desc']
        except:
            print('problem with price_desc descr')
    #for r_type_amount
    if 'r_type_amount' in df.columns:
        try:
            df[['type', 'rooms_amount']] = df['r_type_amount'].str.split(' ∙ ', expand=True)
            del df['r_type_amount']
            df.loc[df['type'] == 'Garasje/Parker', ['type']] = 'Garasje/Parkering'
        except:
            print('problem with r_type_amount descr')
    if 'square_metre' in df.columns:
        try:
            df['square_metre'] = df['square_metre'].str.split().str[0]
            df['square_metre'] = df['square_metre'].map(int)
        except:
            print('problem with square_metre descr')

    return df
```
